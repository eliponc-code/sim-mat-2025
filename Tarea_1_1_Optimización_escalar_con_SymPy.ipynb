# Optimización escalar con SymPy: criterio conocido

Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.

1. $f(x)=x+\frac{1}{x^2}$

```python
import sympy as sym

x = sym.var('x')   #definir var simbolica
f = x+1/x**2 #función

f_der = sym.diff(f, x) #derivada respecto a x
print("f'(x)=", f_der)

criticos = sym.solve(f_der, x) #p. criticos
print("Puntos críticos:", criticos)

f_seg = sym.diff(f, x, 2)
print("f''(x) =", f_seg)

for c in criticos:         #segunda der en p criticos
    seg = f_der.subs(x, c)
    print(f"f''({c}) =", seg)

sym.plot(f, (x, -5, 5), ylim=(-10,10))  #grafica
```

4. $f(x)=\frac{15x}{2}\sqrt{\frac{x}{x-30}}$

```python
import sympy as sym

x = sym.var('x', real=True)
f = 15*x/2 * sym.sqrt(x/(x-30))
print("f(x) =", f)

f_der = sym.diff(f, x)
print("f'(x) =", f_der)

criticos = sym.solve(f_der, x)
print("Puntos críticos:", criticos)

f_der = sym.diff(f, x, 2)
print("f''(x) =", f_der)

for c in criticos:
    try:
        seg = f_der.subs(x, c)
        print(f"f''({c}) =", seg)
        if seg > 0:
            print(f"Mínimo en x = {c} ")
        elif seg < 0:
            print(f"Máximo en x = {c} ")
        else:
            print(f"Inconcluso en x = {c} ")
    except:
        print(f"f''({c}) no definido")

sym.plot(f, (x, 30.01, 100), ylim=(0, 2000))
```

7. $f(x)=100(1+\cos(x)\sin(x))$, para $x\in[0.\frac{\pi}{2}]$

```python
import sympy as sym

x = sym.var('x')
f = 100*(1 + sym.cos(x)*sym.sin(x))
print("f(x) =", f)

f_der = sym.diff(f, x)
print("f'(x) =", f_der)

criticos = sym.solve(f_der, x)
print("Puntos críticos:", criticos)

f_der = sym.diff(f, x, 2)
print("f''(x) =", f_der)

for c in criticos:
    seg = f_der.subs(x, c)
    print(f"f''({c}) =", seg)

sym.plot(f, (x, 0, sym.pi/2), ylim=(-40,150))
```

```python
import sympy as sym

# Definir variable simbólica
x = sym.var('x', real=True)

# Definir la función
f = 100*(1 + sym.cos(x)*sym.sin(x))
print("f(x) =", f)

# Primera derivada
f_der = sym.diff(f, x)
print("f'(x) =", f_der)

# Encontrar puntos críticos (f'(x)=0)
criticos = sym.solve(f_der, x)
print("Puntos críticos (simples y dentro del intervalo):", criticos)

# Segunda derivada
f_second = sym.diff(f, x, 2)
print("f''(x) =", f_second)

# Clasificar los puntos críticos con la segunda derivada
for c in criticos:
    # Solo considerar los que están dentro del intervalo [0, pi/2]
    if 0 <= float(c.evalf()) <= float(sym.pi/2):
        seg = f_second.subs(x, c)
        print(f"f''({c}) =", seg)
        if seg > 0:
            print(f"En x = {c} hay un MÍNIMO local")
        elif seg < 0:
            print(f"En x = {c} hay un MÁXIMO local")
        else:
            print(f"En x = {c} el criterio es INCONCLUSO")

# Evaluar los extremos del intervalo
extremos = [0, sym.pi/2]
for e in extremos:
    val = f.subs(x, e)
    print(f"f({e}) = {val}")

# Graficar la función en [0, pi/2]
sym.plot(f, (x, 0, sym.pi/2), ylim=(0,155))
```
