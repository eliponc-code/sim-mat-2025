1. El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

b = sym.Symbol('b')
l = (10 - b) / 2
h = sym.sqrt(l**2 - (b/2)**2)
V = sym.pi * h**2 * b / 3

b2 = sym.solve(sym.diff(V, b), b)[0]

l2 = (10 - b2) / 2
h2 = h.subs(b, b2)
V2 = V.subs(b, b2)

print(float(b2))
print(float(l2))
print(float(h2))
print(float(V2))

b3 = np.linspace(0.1, 5, 200)
V3 = [V.subs(b, val) for val in b3]

plt.plot(b3, V3)
plt.grid()
plt.show()

```

2. Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

b = sym.Symbol('b', positive=True)
h = (6 - b) / 2
A = b * h

dA = sym.diff(A, b)
b2 = sym.solve(sym.Eq(dA, 0), b)[0]

h2 = h.subs(b, b2)
A2 = A.subs(b, b2)

print("b =", float(b2))
print("h =", float(h2))
print("A =", float(A2))

b3 = np.linspace(0.1, 5.9, 200)
A3 = [A.subs(b, val) for val in b3]

plt.plot(b3, A3)
plt.grid(True)
plt.show()

```
